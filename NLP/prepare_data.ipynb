```python
import pandas as pd
import numpy as np
import urllib.request
import os
import re
```

```python
PATH = "../data/aihub/"

list_dir = os.listdir(f"{PATH}")
list_dir
```

```python
final = []
for dir in list_dir:
    if dir == ".DS_Store":
        continue
    data_list = os.listdir(f"{PATH}/{dir}")
    for data in data_list:
        with open(f'{PATH}/{dir}/{data}', "r") as f:
            s = f.read()
            s = s.split("\n")
            s.append(None)
            final.extend(s)
```

```python
final_df = pd.DataFrame(final)
final_df
```

```python
final_df[0] = final_df[0].map(lambda x: x[4:] if x != None else None)
final_df
```

```python
answer = final_df[0].tolist()
answer.append(None)
answer.pop(0)
```

```python
answer = pd.Series(answer)
answer
```

```python
final_df["A"] = answer
final_df
```

```python
final_df = final_df.rename(columns={0:"Q"})
final_df
```

```python
final_df = final_df.dropna()
final_df
```
